#### 3 Outlier

##### Hapus outlier

```python
#2 Memilih data tanpa outlier
def filter_outlier(df):
  numeric_features = df.select_dtypes(include=['int64', 'float64']).columns
  Q1 = df[numeric_features].quantile(0.25)
  Q3 = df[numeric_features].quantile(0.75)
  IQR = Q3 - Q1
  # Filter dataframe untuk hanya menyimpan baris yang tidak mengandung outliers pada kolom numerik
  condition = ~((df[numeric_features] < (Q1 - 1.5 * IQR)) | (df[numeric_features] > (Q3 + 1.5 * IQR))).any(axis=1)
  df_filtered_numeric = df.loc[condition, numeric_features]

  # Menggabungkan kembali dengan kolom kategorikal
  categorical_features = df.select_dtypes(include=['object']).columns
  df = pd.concat([df_filtered_numeric, df.loc[condition, categorical_features]], axis=1)
  return df
```

```python
# Boxplot sebelum filtering
visual_boxplot(df_clean)
```

```python
# Boxplot sesudah filtering
df_filtered = filter_outlier(df_clean)
visual_boxplot(df_filtered)
```

##### Mengisi data outlier dengan rentang data terdekat (capping)

```python
# Daftar kolom numerik yang ingin ditangani outlier-nya
cols = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']

# Loop setiap kolom
for col in cols:
    Q1 = df_filtered[col].quantile(0.25)
    Q3 = df_filtered[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Capping nilai yang di luar batas
    df_filtered[col] = df_filtered[col].apply(
        lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x
    )
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana tren kualitas udara dari tahun ke tahun

```python
# Visualisasi satu variabel
def visual_line(df,kolom,skala):
  df['tanggal'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
  if skala== 'D':
    df.set_index('tanggal').resample(skala)[kolom].mean().plot(figsize=(12, 4), title=f'Tren Harian {kolom}')
  elif skala == 'M':
    df.set_index('tanggal').resample(skala)[kolom].mean().plot(figsize=(12, 4), title=f'Tren Bulanan {kolom}')
  elif skala == 'Y':
    df.set_index('tanggal').resample(skala)[kolom].mean().plot(figsize=(12, 4), title=f'Tren Tahunan {kolom}')
```

```python
visual_line(df_filtered,'PM2.5','M')
```

```python
visual_line(df_filtered,'PM2.5','Y')
```

```python
def visual_line2(df, faktor, skala='mingguan'):
    # Daftar kolom data yang tersedia
    polutan = ['PM2.5', 'PM10', 'SO2', 'NO2', 'O3']
    cuaca = ['TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']

    # Validasi input faktor
    if faktor == 'polutan':
        data = [col for col in polutan if col in df.columns]
        judul_faktor = 'Polutan Udara'
    elif faktor == 'cuaca':
        data = [col for col in cuaca if col in df.columns]
        judul_faktor = 'Cuaca'
    else:
        raise ValueError("Faktor harus 'polutan' atau 'cuaca'")

    df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
    df.set_index('datetime', inplace=True)

    # Resampling data
    if skala == 'mingguan':
        dff = df[data].resample('W').mean()
        skala_title = 'Mingguan'
    elif skala == 'harian':
        dff = df[data].resample('D').mean()
        skala_title = 'Harian'
    elif skala == 'bulanan':
        dff = df[data].resample('M').mean()
        skala_title = 'Bulanan'
    elif skala == 'tahunan':
        dff = df[data].resample('Y').mean()
        skala_title = 'Tahunan'
    else:
        raise ValueError("Skala harus 'harian', 'mingguan', atau 'bulanan'")

    # Plot data
    plt.figure(figsize=(15, 6))
    for col in data:
        plt.plot(dff.index, dff[col], label=col)

    plt.title(f'Tren {skala_title} {judul_faktor} di Changping')
    plt.xlabel('Waktu')
    plt.ylabel('Konsentrasi (μg/m³ atau ppm)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
visual_line2(df, faktor='polutan', skala='bulanan')
```

```python
visual_line2(df, faktor='polutan', skala='tahunan')
```

### Insight

- Secara tahunan, kebanyakan polutan terus mengalami penurunan mulai dari januari 2015 hingga januari 2017 dan mengalami kenaikan lagi setelah januari 2017.


### Pertanyaan 2: Kapan Hari Berbahaya untuk Aktivitas Luar Ruangan

```python
def hari_berbahaya(df, tahun, threshold=100):
    # Ubah ke format tanggal
    df['datetime'] = pd.to_datetime(df[['year', 'month', 'day']])

    # Filter untuk tahun tertentu
    df_year = df[df['year'] == tahun]

    # Hitung rata-rata PM2.5 harian
    daily_avg = df_year.groupby('date')['PM2.5'].mean()

    # Hari berbahaya: PM2.5 > threshold
    danger_days = daily_avg[daily_avg > threshold]

    print(f"Jumlah hari berbahaya pada tahun {tahun}: {len(danger_days)}")
    print(danger_days.head())

    # Visualisasi
    plt.figure(figsize=(15, 5))
    plt.plot(daily_avg.index, daily_avg.values, label='PM2.5 Harian')
    plt.axhline(threshold, color='red', linestyle='--', label=f'Ambang Batas Bahaya ({threshold})')

    plt.title(f'Hari-Hari dengan PM2.5 Tinggi di Changping - Tahun {tahun}')
    plt.xlabel('Tanggal')
    plt.ylabel('PM2.5 (µg/m³)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
hari_berbahaya(df_filtered, 2013)
hari_berbahaya(df_filtered, 2014)
hari_berbahaya(df_filtered, 2015)
hari_berbahaya(df_filtered, 2016)
hari_berbahaya(df_filtered, 2017)
```

### **Insight:**
- Hari berbahaya karena intensitas PM2.5 tinggi cenderung terjadi hampir sepanjang tahun (mulai februari - desember) dan turun lagi saat januari pada tahun 2014.
- Sepanjang tahun 2014, terdapat 55 hari yang tergolong berbahaya untuk beraktivitas dikarenakan intensitas PM2.5 yang tinggi
- Tidak banyak hari berbahaya pada tahun 2017 karena setelah filtering, tahun 2017 hanya memiliki data hingga maret saja
